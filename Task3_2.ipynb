```python
import pandas as pd

# Load the dataset
from google.colab import files
uploaded = files.upload()

import os
print(os.listdir())  # List files in the current directory

import pandas as pd
cars = pd.read_csv('cars.csv')

# Find the most expensive car company
most_expensive_company = cars.loc[cars['price'].idxmax(), 'company']
print("Most expensive car company:", most_expensive_company)

toyota_cars = cars[cars['company'] == 'toyota']
print("All Toyota cars information:")
print(toyota_cars)

cars_per_company = cars['company'].value_counts()
print("Total number of cars per company:")
print(cars_per_company)

highest_price_per_company = cars.groupby('company')['price'].max()
print("Highest price car for each company:")
print(highest_price_per_company)

average_mileage_per_company = cars.groupby('company')['average-mileage'].mean()
print("Average mileage of each car-making company:")
print(average_mileage_per_company)

cars_sorted_by_price = cars.sort_values(by='price', ascending=False)
print("Cars sorted by price:")
print(cars_sorted_by_price)
```
