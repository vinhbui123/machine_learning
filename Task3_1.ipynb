```python
import pandas as pd

# Load the dataset
from google.colab import files
uploaded = files.upload()

import os
print(os.listdir())  # List files in the current directory

import pandas as pd
users = pd.read_csv('users.csv')

# Get the last 10 entries
last_10_entries = users.tail(10)
print(last_10_entries)

num_observations, num_attributes = users.shape
print(f"Number of observations: {num_observations}")
print(f"Number of attributes: {num_attributes}")

column_names = users.columns
print("Column names:", column_names)

data_types = users.dtypes
print("Data types of each column:")
print(data_types)

unique_occupations = users['occupation'].unique()
print("Unique occupations:", unique_occupations)

most_frequent_occupation = users['occupation'].mode()[0]
print("Most frequent occupation:", most_frequent_occupation)

summary = users.describe(include='all')
print("Summary of all attributes:")
print(summary)

mean_age_per_occupation = users.groupby('occupation')['age'].mean()
print("Mean age per occupation:")
print(mean_age_per_occupation)

# Calculate the male ratio per occupation
male_ratio = users[users['gender'] == 'M'].groupby('occupation').size() / users.groupby('occupation').size()
male_ratio_sorted = male_ratio.sort_values(ascending=False)
print("Male ratio per occupation (sorted):")
print(male_ratio_sorted)

age_min_max = users.groupby('occupation')['age'].agg(['min', 'max'])
print("Minimum and maximum ages per occupation:")
print(age_min_max)

mean_age_occupation_gender = users.groupby(['occupation', 'gender'])['age'].mean()
print("Mean age for each combination of occupation and gender:")
print(mean_age_occupation_gender)

mean_age_occupation_gender = users.groupby(['occupation', 'gender'])['age'].mean()
print("Mean age for each combination of occupation and gender:")
print(mean_age_occupation_gender)

# Calculate the percentage of men and women per occupation
gender_percentage = users.groupby(['occupation', 'gender']).size() / users.groupby('occupation').size() * 100
gender_percentage = gender_percentage.unstack()
print("Percentage of women and men per occupation:")
print(gender_percentage)
```
